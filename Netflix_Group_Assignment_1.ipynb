# Importing the correct data and dropping columns 

```python
import pandas as pd
import numpy as np

df =pd.read_csv('phone_user_review_file_1.csv',encoding='ISO-8859-1')
df2=pd.read_csv('phone_user_review_file_2.csv',encoding='ISO-8859-1')
df3=pd.read_csv('phone_user_review_file_3.csv',encoding='ISO-8859-1')
df4=pd.read_csv('phone_user_review_file_4.csv',encoding='ISO-8859-1')
df5=pd.read_csv('phone_user_review_file_5.csv',encoding='ISO-8859-1')
df6=pd.read_csv('phone_user_review_file_6.csv',encoding='ISO-8859-1')
df7=pd.concat([df,df2,df3,df4,df5,df6], ignore_index=True)
df7=df7.drop(columns=['extract', 'score_max'])
```

# Creating anonymous for the authors that are empty

```python
df7['author']=df7['author'].replace(np.nan,'anonymous',regex=True)# creating anonymous for the authors that are empty
df7
```
