```python
import pandas as pd
import glob
from pathlib import Path

# get current working directory
cwd = Path().absolute()

# get all the files to clean
files = glob.glob(f'{cwd}/resources/*.csv')

# create a new directory to store the cleaned files
cleaned_dir = Path(cwd, 'resources', 'cleaned')

Path.mkdir(cleaned_dir, exist_ok=True)

for file in files:
    df = pd.read_csv(file, encoding='ISO-8859-1')
    df = df.drop(columns=['extract', 'score_max'])
    df['author'] = df['author'].fillna('anonymous')
    df = df.dropna(subset=['score'])
    df = df[df['source'].apply(str.isascii)]
    df = df[df['author'].apply(str.isascii)]
    
    cleaned_file = Path(cleaned_dir, f'cleaned_{Path(file).name}')
    print(f"Finished cleaning! saving file to: {cleaned_file}")
    df.to_csv(cleaned_file, index=False)
```

## Load in cleaned phone review data

```python
df1 = pd.read_csv('resources/cleaned/cleaned_phone_user_review_file_1.csv',encoding='ISO-8859-1')
df2 = pd.read_csv('resources/cleaned/cleaned_phone_user_review_file_2.csv',encoding='ISO-8859-1')
df3 = pd.read_csv('resources/cleaned/cleaned_phone_user_review_file_3.csv',encoding='ISO-8859-1')
df4 = pd.read_csv('resources/cleaned/cleaned_phone_user_review_file_4.csv',encoding='ISO-8859-1')
df5 = pd.read_csv('resources/cleaned/cleaned_phone_user_review_file_5.csv',encoding='ISO-8859-1')
df6 = pd.read_csv('resources/cleaned/cleaned_phone_user_review_file_6.csv',encoding='ISO-8859-1') 
```

## Transform the data so that it corresponds to phone review ERD

```python
from pathlib import Path
previous_range = 0

dfs = [df1, df2, df3, df4, df5, df6]

Path("resources/db_files").mkdir(exist_ok=True)

# Generate a review_id column to be joined on
for i, df in enumerate(dfs):
    review_df = df.drop(columns=["phone_url", "domain"])
    
    dfs[i]['review_id'] = df.index + previous_range

    review_df['review_id'] = df.index + previous_range
    previous_range += len(df.index)
    review_df.to_csv(f'resources/db_files/reviews_{i + 1}.csv', index=False)

# remove comment to sanity check
dfs[0]
```

```python
import uuid

phone_dfs = []
previous_range = 0

# add a new url_id column and create a review_id column to join on
for i, df in enumerate(dfs):
    phone_df = pd.DataFrame().assign(phone_url = df['phone_url'], domain = df['domain'], review_id = df['review_id'])
    phone_df['url_id'] = [uuid.uuid4().hex for i in range(len(df.index))]
    
    phone_df.to_csv(f'resources/db_files/url_{i + 1}.csv', index=False)
    phone_dfs.append(phone_df)

# remove comment to sanity check
phone_dfs[0]
```

```python
%%file test_clean_data.py
import pytest
import pandas as pd
from pathlib import Path
import glob

@pytest.fixture
def test_files():
    cwd = Path().absolute()
    files = glob.glob(f'{cwd}/cleaned/*')
    return files

def test_column_count(test_files):
    for file in test_files:
        df = pd.read_csv(file, encoding='ISO-8859-1')
        assert len(df.columns) == 9

def test_author_null(test_files):
    for file in test_files:
        df = pd.read_csv(file, encoding='ISO-8859-1')
        #check if each author is not empty
        for author in df['author']:
            assert not pd.isna(author)
        
def test_score_nan(test_files):
    for file in test_files:
        df = pd.read_csv(file, encoding='ISO-8859-1')
        #check if each score is not NaN
        for score in df['score']:
            assert not pd.isna(score)
        
def test_rows_dropped(test_files):
    for file in test_files:
        df_clean_row = pd.read_csv(file, encoding='ISO-8859-1').shape[0]
        df_orig_row = pd.read_csv(Path().absolute() / 'resources' / Path(file).name.replace('cleaned_', ''), encoding='ISO-8859-1').shape[0]
        #check if the number of rows are less than original
        assert (df_orig_row != df_clean_row)
    
```
